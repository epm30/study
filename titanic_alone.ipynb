## 2. Exploratory Data Analysis (EDA, 탐색적 데이터 분석)

### Pclass (Class)
- 1,2,3 의 값을 갖는 Ordinal, 서수형 데이터, 순서있음
- Pclass의 따른 생존율 차이 살펴보기
  - pandas의 group by 사용

```
# 3클래스의 생존 비율이 현저히 떨어지므로 1클래스의 생존율이 훨씬 높다.
sns.countplot(data = df_train, x = "Pclass", hue = "Survived")
```

```
# Pclass 칼럼의 항목 갯수
df_train[['Pclass','Survived']].groupby("Pclass", as_index = True).count()
```

```
# Pclass별 생존자 수
# P1(136/216), P2(87/184), P3(119/491)
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=True).sum()
```

```
pd.pivot_table(df_train, index = "Pclass", values = "Survived")
```

```
# crosstab 으로 확인
pd.crosstab(df_train['Pclass'], df_train['Survived'], 
            margins=True).style.background_gradient(cmap='summer_r')
```

```
pd.crosstab(df_train["Sex"], df_train["Survived"],
                                         margins = True).style.background_gradient(cmap = 'summer_r')
```

```
# 클래스별 생존률
# P1 : (136 / (80_136)) ==> 63%
df_train[["Pclass", 'Survived']].groupby(['Pclass'],
                                         )
```

```
# 클래스별 생존률
# P1 : (136 / (80+136)) => 63%
df_train[['Pclass', 'Survived']].groupby(['Pclass'], 
             as_index=True).mean().sort_values(by='Survived', 
                                   ascending=False).plot.bar()
```

```
df_train[['Pclass', 'Survived']].groupby(['Pclass'], 
             as_index=True).mean()
```

```
pd.pivot_table(df_train, index = "Pclass", values = "Survived").sort_values(by = 'Survived', ascending = False).plot.bar()
```

- Pclass가 좋을 수록 생존률이 높음





```
# label 에 따른 갯수 확인
y_position = 1.02
f, ax = plt.subplots(1, 2, figsize = (18, 8))
df_train["Pclass"].value_counts().plot.bar(color = ['#CD7F32','#FFDF00','#D3D3D3'], ax = ax[0])
ax[0].set_title('Number of Passengers By Pclass', y = y_position)
ax[0].set_ylabel('Count')
sns.countplot('Pclass', hue = 'Survived', data = df_train, ax = ax[1])
ax[1].set_title('Pclass : Survived vsv Dead', y = y_position)
plt.show()
```

- 클래스가 높을 수록(1st class) 생존 확률이 높음
  - P1(63%), P2(48%), P3(25%)
- 생존에 Pclass가 큰 영향을 미친다고 생각할 수 있음
  - 모델을 만들 때, Pclass feature를 사용하는 것이 좋을 것이라 판단할 수 있음

### 성별(Sex)
- 성별로 생존률 확인
- pandas groupby와 seaborn countplot 사용


```
f, ax = plt.subplots(1, 2, figsize = (18,8))
df_train[["Sex", "Survived"]].groupby(["Sex"], as_index = True).mean().plot.bar(ax = ax[0])
ax[0].set_title('Survived vs Sex')
sns.countplot(data = df_train, x = "Sex", hue = "Survived", ax = ax[1])
ax[1].set_title('Sex : Survived vs Dead')
plt.show()
```

- female(여성)이 생존할 확률이 높음

### Pclass 와 Sex
- Pclass와 Sex 두 feature에 대한 생존 확률 확인

```
# 3개의 차원 데이터로 이루어진 그래프 그리기
sns.factorplot('Pclass', 'Survived', hue = 'Sex', data = df_train, size = 6, aspect = 1.5)
pass
```

- 모든 클래스에서 남성보다 여성의 생존율이 월등히 높다.
- 남여 구분 없이 클래스가 낮을 수록(등급이 높을수록) 생존 확률이 높음

```
# cloumn 대신 hue 사용
# col사용시 값에 따라 개별 그래프로 작성
# hue를 사용시 한개에 여러개의 그래프로 작성
sns.factorplot(x='Sex', y='Survived', col='Pclass',
              data=df_train, satureation=.5,
               size=9, aspect=1)
sns.factorplot(x='Sex', y='Survived', hue='Pclass',
              data=df_train, satureation=.5,
               size=9, aspect=1)
```

### Age
- 나이에 따른 생존 확률 확인

```
# 간단한 통계 보기
  print('제일 나이 많은 탑승객 : {:.1f} Years'.format(df_train['Age'].max()))
  
  print('제일 나이 어린 탑승객 : {:.1f} Years'.format(df_train['Age'].min()))
  
  print('탑승객 평균 나이 : {:.1f} Years'.format(df_train['Age'].mean()))
```

```
# 생존에 따른 Age의 히스토그램
# kdeplot()
fig, ax = plt.subplots(1, 1, figsize = (9, 5))
sns.kdeplot(df_train[df_train["Survived"] == 1]['Age'], ax = ax)
sns.kdeplot(df_train[df_train["Survived"] == 0]["Age"], ax = ax)
plt.legend(["Survived == 1", 'Survived == 0'])
plt.show()
```

- 나이가 어릴수록 생존률이 높은 모습을 보인다.
